## Задача №1

### Часть 1

Построить в общих осях графики для:

- $\alpha = 1, \beta = 1$

- $\alpha = 2, \beta = 1$

- $\alpha = 1, \beta = 2$

```python
import  matplotlib.pyplot  as  plt
import numpy as np
# Не падает - значит ошибок не было :)
np.seterr(divide='ignore')
%matplotlib inline

def f(x, a, b):
    return (x**b + a**b) / x**b

x = np.around(np.arange(-5, 5, 0.01), 2)

fig, axes = plt.subplots(figsize=(13,8), dpi=120)

axes.set_xlabel('x', fontsize = 16)
axes.set_ylabel('f(x)', fontsize = 16)
axes.set_title(r'График функции $f(x) = \frac{x^{\beta} + \alpha^{\beta}}{x^{\beta} }$ '+\
               r'при различных $\alpha$ и $\beta$', fontsize = 16)

axes.set_ylim(-30, 30)
axes.plot(x, f(x, 1, 1), label=r'$\alpha = 1, \beta = 1$')
axes.plot(x, f(x, 2, 1), label=r'$\alpha = 2, \beta = 1$')
axes.plot(x, f(x, 1, 2), label=r'$\alpha = 1, \beta = 2$')
axes.legend(loc = 1)

plt.savefig('График.svg')
```

<b>Вывод:</b><br>
При $\alpha = 1, \beta = 1$ и $\alpha = 2, \beta = 1$ график представляет собой ветви гиперболы.
При $\alpha = 1, \beta = 2$ левая ветвь гиперболы отображается во вторую координатную четверть, так как каждый член уравнения, задающего функцию, возводится в степень $\beta = 2$.

### Часть 2

Построить в общих осях графики для $x>0$.

На том же графике сделать 2 врезки, демонстрирующие поведение графиков на 2 интервалах:

- для малых x

- для больших x

Необходимо продемонстрировать возможность (или невозможность) пересечений и стремление функций.

Цвет линий на врезках и основном графике должен быть одинаковым для одних и тех же значений $\alpha$ и $\beta$.

```python
import  matplotlib.pyplot  as  plt
import numpy as np
np.seterr(divide='ignore')
%matplotlib inline

def f(x, a, b):
    return (x**b + a**b) / x**b

x = np.arange(0, 30, 0.01)

fig, axes = plt.subplots(figsize=(13,8), dpi=120)

# Главные оси
axes.set_xlabel('x', fontsize = 16)
axes.set_ylabel('f(x)', fontsize = 16)
axes.set_title(r'График функции $f(x) = \frac{x^{\beta} + \alpha^{\beta}}{x^{\beta} }$ при различных $\alpha$ и $\beta$', fontsize = 16)
axes.set_ylim(0, 30)
axes.plot(x, f(x, 1, 1), label=r'$\alpha = 1, \beta = 1$')
axes.plot(x, f(x, 2, 1), label=r'$\alpha = 2, \beta = 1$')
axes.plot(x, f(x, 1, 2), label=r'$\alpha = 1, \beta = 2$')
axes.legend(loc = 1)

# Окошко "при больших x"
x = np.arange(0, 1e3, 0.01)
axes1 = fig.add_axes([0.6, 0.2, 0.25, 0.25])
axes1.set_xlim(1e3 - 10, 1e3)
axes1.set_ylim(0.999, 1.003)
axes1.plot(x, f(x, 1, 1), label=r'$\alpha = 1, \beta = 1$')
axes1.plot(x, f(x, 2, 1), label=r'$\alpha = 2, \beta = 1$')
axes1.plot(x, f(x, 1, 2), label=r'$\alpha = 1, \beta = 2$')
axes1.set_title(r'График функции f(x) при больших значениях x', fontsize = 10)

# Окошко "при малых x"
x = np.arange(0, 0.05, 0.001)
axes2 = fig.add_axes([0.25, 0.55, 0.25, 0.25])
axes2.set_xlim(-0.001, 0.035)
axes2.set_ylim(1e3 - 10, 1e3)
axes2.plot(x, f(x, 1, 1), label=r'$\alpha = 1, \beta = 1$')
axes2.plot(x, f(x, 2, 1), label=r'$\alpha = 2, \beta = 1$')
axes2.plot(x, f(x, 1, 2), label=r'$\alpha = 1, \beta = 2$')
axes2.set_title(r'График функции f(x) при малых значениях x', fontsize = 10)

plt.savefig('График_1.svg')
```

### Часть 3

Построить в общих осях графики для $x<0$.

На том же графике сделать 1 врезку, демонстрирующую поведение графиков при удалении x от 0 к $-\infty$.

Необходимо продемонстрировать возможность (или невозможность) пересечений и стремление функций. Так же нанесите на графики прямую f(x) = 0.

Цвет линий на врезках и основном графике должен быть одинаковым для одних и тех же значений $\alpha$ и $\beta$.

```python
import  matplotlib.pyplot  as  plt
import numpy as np
np.seterr(divide='ignore')
%matplotlib inline

def f(x, a, b):
    return (x**b + a**b) / x**b

x = np.arange(-30, 0, 0.001)

fig, axes = plt.subplots(figsize=(13,8), dpi=120)

# Главные оси
axes.set_xlabel('x', fontsize = 16)
axes.set_ylabel('f(x)', fontsize = 16)
axes.set_title(r'График функции $f(x) = \frac{x^{\beta} + \alpha^{\beta}}{x^{\beta} }$ при различных $\alpha$ и $\beta$', fontsize = 16)
axes.set_ylim(-30, 30)
axes.plot(x, f(x, 1, 1), label=r'$\alpha = 1, \beta = 1$')
axes.plot(x, f(x, 2, 1), label=r'$\alpha = 2, \beta = 1$')
axes.plot(x, f(x, 1, 2), label=r'$\alpha = 1, \beta = 2$')
axes.axhline(0, xmin=0.04, xmax=0.96, color='r', ls='--', label=r'x = 0')
axes.legend()

# Окошко "при стремлении x к минус бесконечности"
x = np.arange(-1e3, 0, 0.01)
axes1 = fig.add_axes([0.2, 0.18, 0.35, 0.24])
axes1.set_xlim(-1e3, 0)
axes1.set_ylim(0.99, 1.01)
axes1.plot(x, f(x, 1, 1), label=r'$\alpha = 1, \beta = 1$')
axes1.plot(x, f(x, 2, 1), label=r'$\alpha = 2, \beta = 1$')
axes1.plot(x, f(x, 1, 2), label=r'$\alpha = 1, \beta = 2$')
labels = [item for item in axes1.get_xticks().tolist()]
labels[0] = r'-$\infty$'
axes1.set_xticklabels(labels) 
axes1.set_title(r'График функции f(x) при стремлении x к -$\infty$', fontsize = 10)

plt.savefig('График_2.svg')
```

- $\alpha = 1, \beta = -1$

Сделайте выводы о поведении графиков

```python
import  matplotlib.pyplot  as  plt
import numpy as np
np.seterr(divide='ignore', invalid='ignore')
%matplotlib inline

def f(x, a, b):
    return (x**b + a**b) / x**b

x = np.around(np.arange(-10, 10, 0.01), 2)

fig, axes = plt.subplots(figsize=(13,8), dpi=120)

axes.set_xlabel('x', fontsize = 16)
axes.set_ylabel('f(x)', fontsize = 16)
axes.set_title(r'График функции $f(x) = \frac{x^{\beta} + \alpha^{\beta}}{x^{\beta} }$ '+\
               r'при различных $\alpha$ и $\beta$', fontsize = 16)

axes.set_ylim(-10, 10)
axes.plot(x, f(x, 1, 0), label=r'$\alpha = 1, \beta = 0$', lw=4)
axes.plot(x, f(x, 0, 0), '-.', label=r'$\alpha = 0, \beta = 0$')
axes.plot(x, f(x, 0, 1), label=r'$\alpha = 0, \beta = 1$', lw=4)
axes.plot(x, f(x, 0, 2), '--', label=r'$\alpha = 0, \beta = 2$')
axes.plot(x, f(x, 1, -1), label=r'$\alpha = 1, \beta = -1$')
axes.plot(x, f(x, 2, -1), label=r'$\alpha = 2, \beta = -1$')
axes.legend()

plt.savefig('График_3.svg')
```

## Задача №3

Построить в общих осях графики для:

- $\alpha = 1, \beta = 0.5$

- $\alpha = 1, \beta = -0.5$

- $\alpha = 1, \beta = -1.5$

Сделайте выводы о поведении графиков, включая возрастание/убывание и выпуклость/вогнутость

```python
import  matplotlib.pyplot  as  plt
import numpy as np
np.seterr(divide='ignore', invalid='ignore')
%matplotlib inline

def f(x, a, b):
    return (x**b + a**b) / x**b

x = np.around(np.arange(0, 10, 0.01), 2)

fig, axes = plt.subplots(figsize=(13,8), dpi=120)

axes.set_xlabel('x', fontsize = 16)
axes.set_ylabel('f(x)', fontsize = 16)
axes.set_title(r'График функции $f(x) = \frac{x^{\beta} + \alpha^{\beta}}{x^{\beta} }$ '+\
               r'при различных $\alpha$ и $\beta$', fontsize = 16)

axes.set_ylim(0, 10)
axes.plot(x, f(x, 1, 0.5), label=r'$\alpha = 1, \beta = 0.5$')
axes.plot(x, f(x, 1, -0.5), label=r'$\alpha = 1, \beta = -0.5$')
axes.plot(x, f(x, 1, -1.5), label=r'$\alpha = 1, \beta = -1.5$')
axes.legend(loc = 1)

plt.savefig('График_4.svg')
```

<b>Вывод:</b><br>
График функции при $\alpha = 1, \beta = 0.5$ представляет собой ветвь гиперболы и убывает на промежутке от 0 до +$\infty$, вогнутая (выпуклая вниз) на промежутке от 0 до +$\infty$.<br>
График функции при $\alpha = 1, \beta = -0.5$ переходит в график функции вида $k*\sqrt{x} + b$ и возрастает на промежутке от 0 до +$\infty$, выпуклая (выпуклая вверх) на промежутке от 0 до +$\infty$.<br>
График функции при $\alpha = 1, \beta = -1.5$ переходит в график функции вида $k*x^{\alpha} + b$ при $\alpha > 1$ и возрастает на промежутке от 0 до +$\infty$, вогнутая (выпуклая вниз) на промежутке от 0 до +$\infty$.<br>
